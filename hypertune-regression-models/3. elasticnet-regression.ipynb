"Elastic Net is a linear regression technique that combines L1 (Lasso) and L2 (Ridge) regularization. It can both shrink coefficients and perform feature selection, making it effective for high-dimensional data with correlated features."

2. Cost Function

Elastic Net’s loss function is:
Cost=i=1∑n (yi −y^ i )2+λ1 j=1∑p ∣βj ∣+λ2 j=1∑p βj2

Where:
λ1 controls L1 (feature selection)
λ2 controls L2 (shrinkage/stability)
By tuning both, you balance sparsity and stability.

3. When to Use Elastic Net

    High-dimensional datasets with many correlated features

    When Lasso alone struggles (e.g., selects only one feature from a group of correlated features)

    When you want both feature selection and coefficient shrinkage

    Useful for p >> n problems (more features than samples)

4. How It Works

    Combines penalties from Ridge and Lasso.

    Optimizes coefficients to minimize the penalized loss function.

    Keeps important correlated features together (unlike Lasso which may pick only one).

    Hyperparameters λ1 andλ2 are usually tuned via cross-validation.

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

#load data
X,y = load_diabetes(return_X_y=True)

# split train
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)

# Linear Regression
reg = LinearRegression()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

### Apply ElasticNet regression

```python
# ElasticNet
reg = ElasticNet(alpha=0.005,l1_ratio=0.9)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```
